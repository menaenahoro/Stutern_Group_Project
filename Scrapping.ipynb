```python
def Get_info(price, mile, conditions, years, names, location):
    driver = webdriver.Chrome('chromedriver',options=options)
    
    driver.get("https://autochek.africa/ng/cars-for-sale")
    '''Directing webdriver to get the web address'''

     
    k=1
    while k<245:
        drop_search = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[2]/div/div[1]/div[3]")
        '''Directing driver to search for a particular xpath'''
        #The xpath displayed is for the search results of Lagos
        ''' we can use find_elements_by_name, by_xpath, by_id, by_class_name, 
        by_css_selector, by_link_text, by_partial_link_text, 
        find_element_by_name, by_xpath, by_id, by_css_selector, by_class_name, 
        by_link_text, by_tag_name, by_partial_link_text'''


        
        prices = drop_search.find_elements(By.XPATH,"//*[contains(concat( ' ', @class, ' ' ), concat( ' ', 'price', ' ' ))]")
        '''Directing driver to pick the xpath of the prices in the search result'''
        for value in prices:
            price.append(value.text)
        
        
        mileage = drop_search.find_elements(By.XPATH, "//*[contains(concat( ' ', @class, ' ' ), concat( ' ', 'detail-section', ' ' )) and (((count(preceding-sibling::*) + 1) = 1) and parent::*)]//*[contains(concat( ' ', @class, ' ' ), concat( ' ', 'text', ' ' ))]")
        '''Directing driver to pick the xpath of the mileage in the search result'''
        for value in mileage:
            mile.append(value.text)

        
        condition = drop_search.find_elements(By.XPATH,"//*[contains(concat( ' ', @class, ' ' ), concat( ' ', 'detail-section', ' ' )) and (((count(preceding-sibling::*) + 1) = 3) and parent::*)]//*[contains(concat( ' ', @class, ' ' ), concat( ' ', 'text', ' ' ))]")
        '''Directing driver to pick the xpath of the condition in the search result'''
        for value in condition:
            conditions.append(value.text)
        
        
        year  = drop_search.find_elements(By.XPATH, "//*[contains(concat( ' ', @class, ' ' ), concat( ' ', 'year', ' ' ))]")
        '''Directing driver to pick the xpath of the year in the search result'''
        for value in year:
            years.append(value.text)
            
        
        name  = drop_search.find_elements(By.XPATH, "//*[contains(concat( ' ', @class, ' ' ), concat( ' ', 'jsx-3456531073', ' ' )) and contains(concat( ' ', @class, ' ' ), concat( ' ', 'name', ' ' ))]")
        '''Directing driver to pick the xpath of the names in the search result'''
        for value in name:
            names.append(value.text)
            
        
        locations  = drop_search.find_elements(By.XPATH, "//*[contains(concat( ' ', @class, ' ' ), concat( ' ', 'detail-section', ' ' )) and (((count(preceding-sibling::*) + 1) = 2) and parent::*)]//*[contains(concat( ' ', @class, ' ' ), concat( ' ', 'text', ' ' ))]")
        '''Directing driver to pick the xpath of the locations in the search result'''
        for value in locations:
            location.append(value.text)


        
        time.sleep(1)
        
        next_page = driver.find_element(By.XPATH,"//*[contains(concat( ' ', @class, ' ' ), concat( ' ', 'next', ' ' ))]")
        driver.execute_script("arguments[0].click();", next_page)
        '''Directing driver to find the xpath of the next button and click on it'''
        # For going to the next page

        k+=1
    
    driver.quit()
```

```python
price = []
mile = []
conditions = []
years = []
names = []
location = []

Get_info(price, mile, conditions, years, names, location)
```

```python
data = {"Name": names, "Price": price, "Mileage": mile, "Condition": conditions, "Year": years, "locations": location}
print(data)
```

```python
import pandas as pd
Autochek_web_data = pd.DataFrame.from_dict(data)
Autochek_web_data.to_csv('Autochek_web_data.csv')
```
